```python
import folium
from folium import plugins
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_csv('/Users/mille/Desktop/WDFW-Hatcheries.csv')
```

```python
df.head()
```

```python
m = folium.Map([47.9641, -123.1151], zoom_start=4)
m

```

```python
for index, row in df.iterrows():
    folium.Marker([row['Latitude'], row['Longitude']], 
                  popup=row['Facility_Name'],
                  icon=folium.Icon(icon='green')
                 ).add_to(m)
m
m.save('/Users/mille/Washington.html')
```
