```python
import folium
from folium import plugins
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_csv('/Users/mille/Desktop/ACobalt.csv')
```

```python
df.head()
```

```python
m = folium.Map([30, -7.0], zoom_start=4)
m
# Russia lat long 

```

```python
for index, row in df.iterrows():
    folium.Marker([row['latitude'], row['longitude']], 
                  popup=row['location'],
                  icon=folium.Icon(icon='green')
                 ).add_to(m)
m
m.save('/Users/mille/AfricaCobalt.html')
```
